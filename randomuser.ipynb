```python
import requests
from pprint import pprint

# Define the API endpoint and parameters
url = "https://api.randomuser.me/"
params = {
    "results": 5,
    "gender": "male",
    "nat": "us"
}

# Make the GET request
response = requests.get(url, params=params)

# Check status code
print(f"Status Code: {response.status_code}")

if response.status_code == 200:
    # Convert JSON response to Python dictionary
    data = response.json()
    
    # Pretty-print the whole response
    pprint(data)
    
    # Loop through each user and print first name
    for user in data['results']:
        first_name = user['name']['first']
        print(f"First Name: {first_name}")
else:
    print("Failed to retrieve data")

# Extra: Print astronauts currently in space
astro_url = "http://api.open-notify.org/astros.json"
astro_response = requests.get(astro_url)
if astro_response.status_code == 200:
    astro_data = astro_response.json()
    print(f"\nNumber of astronauts in space: {astro_data['number']}")
    for person in astro_data['people']:
        print(person['name'])
else:
    print("Failed to retrieve astronaut data")
```
